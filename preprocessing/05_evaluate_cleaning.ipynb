```python
from database.MongoDBClient import MongoDBClient
from database.Recipe import Recipe
import random
```

```python
# Connect to mongo
mongo = MongoDBClient()
```

```python
# Gather all recipes
recipes = [Recipe.from_dict(doc) for doc in mongo.collection.find()]
print(len(recipes))
```

```python
# Step 2: Group recipes by source
source_dict = {}
for recipe in recipes:
    source = recipe.source
    if source not in source_dict:
        source_dict[source] = []
    source_dict[source].append(recipe)

# Sample 10 recipes from each source
sampled_recipes = {}
for source, recipes in source_dict.items():
    # Randomly sample 10 recipes or all if less than 10
    sampled_recipes[source] = random.sample(recipes, min(10, len(recipes)))

# Print samples
for source, samples in sampled_recipes.items():
    print(f"Samples from source '{source}':")
    for recipe in samples:
        print(recipe)
    print("\n")
```
